```python
def filter_wrapper(df):
    """
    
    """
    df = df[df['holiday'] != 1 ]
    df= df[df['workingday'] == 1]
    df = df[14 <= df['hr']]
    df = df[22 >= df['hr']]
    
    df_yr0 = df[df['yr'] == 0]
    df_yr1 = df[df['yr'] == 1]
    
    df_yr0_spring = df_yr0[df_yr0['season'] == 2]
    df_yr0_summer = df_yr0[df_yr0['season'] == 3]

    df_yr1_spring = df_yr1[df_yr1['season'] == 2]
    df_yr1_summer = df_yr1[df_yr1['season'] == 3]

    return df_yr0_spring, df_yr0_summer, df_yr1_spring, df_yr1_summer

df0_spring, df0_summer, df1_spring, df1_summer = filter_wrapper(df_hour)

#sns.set(rc={'figure.figsize':(15,5)})
sns.lineplot(data=df1_summer, x='hr',y='cnt', hue='weekday')
```

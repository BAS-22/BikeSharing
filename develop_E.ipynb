```python
def create_dataframes():
    return pd.read_csv('day.csv'), pd.read_csv('hour.csv')
    
def season_hour_wrapper(df):
    """
    
    """
    df = df[df['holiday'] != 1 ]
    df = df[df['workingday'] == 1]
    df = df[1 <= df['weekday'] ]
    
    
    df = df[15 <= df['hr']]
    df = df[19 >= df['hr']]
    
    df_yr0 = df[df['yr'] == 0]
    df_yr1 = df[df['yr'] == 1]
    
    df_yr0_spring = df_yr0[df_yr0['season'] == 2]
    df_yr0_summer = df_yr0[df_yr0['season'] == 3]

    df_yr1_spring = df_yr1[df_yr1['season'] == 2]
    df_yr1_summer = df_yr1[df_yr1['season'] == 3]


    return df_yr0_spring, df_yr0_summer, df_yr1_spring, df_yr1_summer

def count_aggregator(df_hr):
    """ 
    
    """
    return df_hr['cnt'].groupby(df_hr['dteday']).sum()

def collect_data():
    """ 
    
    """

    _, df_hour = create_dataframes()
    df0_spring, df0_summer, df1_spring, df1_summer = season_hour_wrapper(df_hour)
    
    frames = {}
    frames['spring0'], frames['summer0'] = df0_spring, df0_summer
    frames['spring1'], frames['summer1'] = df1_spring, df1_summer

    for key in frames.keys():
        frames[key] = count_aggregator(frames[key])
    return frames

def plot_hist(seasons):
    """ 
    
    """
    fig, ax = plt.subplots(2, figsize=(8,10))
    cnt = 0
    for season in ['spring', 'summer']:
        year0 = seasons[f'{season}0']
        year1 = seasons[f'{season}1']
        ax[cnt].hist(x=year0, label='year 0', alpha=0.6, bins=30)
        ax[cnt].hist(x=year1, label='year 1', alpha=0.6, bins=30)
       
        cnt += 1
    ax[0].set_title('Spring')
    ax[1].set_title('Summer')
        
    
    plt.legend()
    plt.show()

def plot_box(seasons):
    """ 
    
    """
    fig, ax = plt.subplots(2, figsize=(8,10))
    cnt = 0
    
    for season in ['spring', 'summer']:
        year0 = seasons[f'{season}0']
        year1 = seasons[f'{season}1']
        ax[cnt].boxplot(x=year0)
        ax[cnt].boxplot(x=year1)
        cnt += 1
    ax[0].set_title('Spring')
    ax[1].set_title('Summer')
        
    
    plt.legend()
    plt.show()

def E(r, n):
    inf, phi, Phi = float('inf'), stat.NormalDist().pdf, stat.NormalDist().cdf
    def f(x):

        F = Phi(x)
        return x*(1-F)**(r-1)*F**(n-r)*phi(x)
    return r*binom(n, r)*quad(f, -inf, inf)[0]

def qq_plot(seasons):
    
    fig, ax = plt.subplots(len(seasons.keys()), figsize = (5,15))
    cnt = 0
    norm_cdf = stat.NormalDist().cdf
    for key in seasons.keys():
        n = len(seasons[key])
        
        data_quantiles = np.array([ E(k+1,n) for k in range(n)])
        data_quantiles.sort()
        normal_quantiles = np.linspace(-3,3, len(data_quantiles))
        
        ax[cnt].scatter(normal_quantiles, data_quantiles, label=key)
        ax[cnt].plot((-3,3),(-3,3), linestyle='--', color='black')
        ax[cnt].set_title(key)
        cnt += 1
    plt.legend()

def empirical_cdf(seasons):
    fig, ax = plt.subplots(len(seasons.keys()), figsize = (5,15))
    cnt = 0

    for key in seasons.keys():
        arr = np.array(seasons[key])
        arr.sort()
        arr = ( arr - np.mean(arr) ) / np.std(arr)
        ax[cnt].plot(arr, np.linspace(0,1, len(arr), endpoint=False))
        cnt += 1


filtered_seasons = collect_data()
plot_hist(filtered_seasons)
plot_box(filtered_seasons)
qq_plot(filtered_seasons)
empirical_cdf(filtered_seasons)
```
